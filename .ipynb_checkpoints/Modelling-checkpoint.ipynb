```python
import pandas as pd
import numpy as ny
import sys
from sklearn.linear_model import LinearRegression

TIME = "time"
Year = "year"
data = pd.read_csv('ireland-merged.csv',  parse_dates=[TIME],  dtype={Year: "int"})
data.head
data.columns
```

```python
#model 1 wihtout date

y = data['load']
x = data[['rain','temp','day_of_week']]
#model = LinearRegression()
#model.fit(x,y)
#same as
lm = LinearRegression().fit(x,y)
print(lm.intercept_)
print(lm.coef_)
```

```python
#rehaping y
#model = LinearRegression().fit(x,y.reshape((-1,1)))
#not working - 'Series' object has no attribute 'reshape'

print(lm.intercept_)
print(lm.coef_)
```

```python
y_predict = lm.predict(x)
print(y_predict)
import matplotlib.pyplot as plt
plt.plot(y_predict,'r', y,'b' )
#plt.rcParams["figure.figsize"] = (50,7)
plt.figure(figsize=(100,50))
plt.show()

#red = predicted
#blue = actual
```

```python
#Spliting data sets to show accuracy of the model

data = data.set_index(data['time'])
train = data['2015-01-01':'2019-06-30']
test  = data['2020-06-30':'2020-12-31']
```

```python
#Spliting data sets

data = data.set_index(data['time'])
train = data['2015-01-01':'2020-03-31']
test  = data['2020-04-01':]
```

```python
train.tail()
```

```python
test.head()
```

```python
#define testing and training data
y_train = train['load']
x_train = train[['rain','temp','day_of_week']]
x_test = test[['rain','temp','day_of_week']]
#y_predict = model.predict(x_test)



y_actual = test[['load']]
y_actual.reset_index(drop=True, inplace=True)
#y_actual = y_actual.reshape(1,62)
print(y_actual)
```

```python
#LINER REGRESSION
#using train data to train
#model 1 wihtout date
#model.fit(x,y)
#same as

lm = LinearRegression().fit(x_train,y_train)
print(lm.intercept_)
print(lm.coef_)
```

```python
#display predictions

y_predict = lm.predict(x_test)
print(y_predict)

print(type(y_predict))
print(type(y_actual))
```

```python
#visalize results 

#LINEAR REGRESSION

import matplotlib.pyplot as plt
plt.plot(y_predict,'r', y_actual[['load']],'b' )
plt.rcParams["figure.figsize"] = (50,7)
plt.figure(figsize=(50,7))
plt.show()

#red = predicted
#blue = actual
```
